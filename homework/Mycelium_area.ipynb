```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
%matplotlib inline
dir_data = './data/'
f_app = os.path.join(dir_data, 'Mycelium_area.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app, encoding='big5')
app_train.rename(columns = {'':'number'})
app_train.head(10)
app_train['label'].value_counts()
plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.scatter(app_train['D4'],(app_train['D5']-app_train['D4']))
plt.scatter(app_train['D4'],(app_train['D6']-app_train['D4']))
plt.scatter(app_train['D4'],(app_train['D7']-app_train['D4']))
plt.scatter(app_train['D4'],(app_train['D8']-app_train['D4']))
plt.subplot(2,2,2)
plt.scatter(app_train['D4'],(app_train['D5']/app_train['D4']))
plt.scatter(app_train['D4'],(app_train['D6']/app_train['D4']))
plt.scatter(app_train['D4'],(app_train['D7']/app_train['D4']))
plt.scatter(app_train['D4'],(app_train['D8']/app_train['D4']))
```

```python
plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.scatter(app_train['NO'],(app_train['D5']-app_train['D4']))
plt.scatter(app_train['NO'],(app_train['D6']-app_train['D4']))
plt.scatter(app_train['NO'],(app_train['D7']-app_train['D4']))
plt.scatter(app_train['NO'],(app_train['D8']-app_train['D4']))
plt.subplot(2,2,2)
plt.scatter(app_train['NO'],(app_train['D5']/app_train['D4']))
plt.scatter(app_train['NO'],(app_train['D6']/app_train['D4']))
plt.scatter(app_train['NO'],(app_train['D7']/app_train['D4']))
plt.scatter(app_train['NO'],(app_train['D8']/app_train['D4']))
```
