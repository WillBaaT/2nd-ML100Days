```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
%matplotlib inline
dir_data = './data/'
f_app = os.path.join(dir_data, 'Mycelium_area.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app, encoding='big5')
app_train.rename(columns = {'':'number'})
app_train.head(10)
app_train['label'].value_counts()
plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.scatter(app_train['D4'],(app_train['D5']-app_train['D4']),label='1')
plt.scatter(app_train['D4'],(app_train['D6']-app_train['D5']),label='2')
plt.scatter(app_train['D4'],(app_train['D7']-app_train['D6']),label='3')
plt.scatter(app_train['D4'],(app_train['D8']-app_train['D7']),label='4')
plt.legend()
plt.subplot(2,2,2)
plt.scatter(app_train['D4'],(app_train['D5']/app_train['D4']),label='1')
plt.scatter(app_train['D4'],(app_train['D6']/app_train['D5']),label='2')
plt.scatter(app_train['D4'],(app_train['D7']/app_train['D6']),label='3')
plt.scatter(app_train['D4'],(app_train['D8']/app_train['D7']),label='4')
plt.legend()
```

```python
plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.scatter((app_train['D5']-app_train['D4']),(app_train['D6']-app_train['D5']))
plt.scatter((app_train['D5']-app_train['D4']),(app_train['D7']-app_train['D6']))
plt.scatter((app_train['D5']-app_train['D4']),(app_train['D8']-app_train['D7']))
plt.subplot(2,2,2)
plt.scatter((app_train['D5']/app_train['D4']),(app_train['D6']/app_train['D5']))
plt.scatter((app_train['D5']/app_train['D4']),(app_train['D7']/app_train['D6']))
plt.scatter((app_train['D5']/app_train['D4']),(app_train['D8']/app_train['D7']))
```

```python
app_train['D5-4']=app_train['D5']-app_train['D4']
app_train['D6-5']=app_train['D6']-app_train['D5']
app_train['D7-6']=app_train['D7']-app_train['D6']
app_train['D8-7']=app_train['D8']-app_train['D7']
app_train['D54']=app_train['D5']/app_train['D4']
app_train['D65']=app_train['D6']/app_train['D5']
app_train['D76']=app_train['D7']/app_train['D6']
app_train['D87']=app_train['D8']/app_train['D7']

peel=app_train.groupby(['label','peel'])
peel.mean()
```

```python
plt_column = ['D5-4','D6-5','D7-6','D8-7','D54','D65','D76','D87']
plt_by = ['label','peel']
i=0
while i<len(plt_column):
    app_train.boxplot(column=plt_column[i], by = plt_by, showfliers = False, figsize=(12,5))
    i=i+1
```

```python
binn = np.linspace(1.2,2.3,12)
print(binn)
app_train['D54_g']=pd.cut(app_train['D54'].values,binn,right=False)
app_train['D65_g']=pd.cut(app_train['D65'].values,binn,right=False)
app_train['D76_g']=pd.cut(app_train['D76'].values,binn,right=False)
app_train['D87_g']=pd.cut(app_train['D87'].values,binn,right=False)
print(app_train['D54_g'].value_counts())
print(app_train['D65_g'].value_counts())
print(app_train['D76_g'].value_counts())
print(app_train['D87_g'].value_counts())
app_train_sorted = app_train['D54_g'].unique().sort_values()

import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(12,12))

for i in range(len(app_train_sorted)):
    plt.subplot(4,2,1)
    sns.distplot(app_train.loc[(app_train['D54_g'] == app_train_sorted[i]), 'D54'], label = str(app_train_sorted[i]))
for i in range(len(app_train_sorted)):
    plt.subplot(4,2,2)
    sns.distplot(app_train.loc[(app_train['D65_g'] == app_train_sorted[i]), 'D65'], label = str(app_train_sorted[i]))
for i in range(len(app_train_sorted)):
    plt.subplot(4,2,3)
    sns.distplot(app_train.loc[(app_train['D76_g'] == app_train_sorted[i]), 'D76'], label = str(app_train_sorted[i]))
for i in range(len(app_train_sorted)):
    plt.subplot(4,2,4)
    sns.distplot(app_train.loc[(app_train['D87_g'] == app_train_sorted[i]), 'D87'], label = str(app_train_sorted[i]))
```
